#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_df_columns = ['CreationDate', 'DisplayName', 'LastAccessDate','WebsiteUrl', 'Location', 'AboutMe','AccountId', 'Age', 'ProfileImageUrl']
posts_df_columns = ['PostTypeId', 'AcceptedAnswerId', 'CreaionDate','Body', 'LasActivityDate', 'Title', 'Tags','AnswerCount', 'FavoriteCount', 'LastEditorUserId','LastEditDate', 'CommunityOwnedDate', 'ParentId', 'ClosedDate','OwnerDisplayName', 'LastEditorDisplayName']

users_df_selected = users_df.drop(users_df_columns,axis=1)
posts_df_selected = posts_df.drop(posts_df_columns,axis=1)
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
estadisticas = data.describe().T
estadisticas
```

```python
estadisticas['IQR'] = estadisticas['75%']-estadisticas['25%']


outliers = pd.DataFrame(columns=data.columns)

for col in estadisticas.index:
    iqr = estadisticas.at[col,'IQR'] # obtenemos el rango intercuantil
    cutoff = iqr*1.5 # Obtenemos nuestro rango intercuartil
    lower = estadisticas.at[col,'25%']-cutoff # obtenemos el minimo
    upper = estadisticas.at[col,'75%'] + cutoff # obtenemos el maximo
    results = data[(data[col]<lower)|(data[col]>upper)].copy() # buscaremos cuales son las columnas tienen un outlier
    results['Outlier'] = col # obtenemos el nombre de la columna que tiene el valor fuera de rango
    outliers = outliers.append(results) # agregamos las columnas al df

outliers
```

```python
outliers['Outlier'].value_counts()
```

```python
outliers.to_csv('Outliers.csv')
```
